```python
FILE_NAME = "students.txt"
import os 
def read_students():
    students = []
    if not os.path.exists(FILE_NAME):
        return students
    with open(FILE_NAME, "r") as file:
        for line in file:
            data = line.strip().split(",") 
            if (len(data)==4):
                student = { "id" : data[0], "name" : data[1], "class" : data[2], "marks" : float(data[3])} 
                students.append(student)
    return student


def write_students(students):
    file = open(FILE_NAME, "w")
    for i in students:
        file.write(f"{i["id"]},{i["name"]}, {i["class"]}, {i["marks"]}\n")
      


def add_student():
    student=read_students()
    sid=input("ID:")
    for s in students:
        if s["id"]==sid:
            print("student with this ID already exists!")
            return
        name = input("Enter Name:")
        marks = float(input("Enter Marks:"))
        students.append({"id": sid, "name": name, "marks": marks})
        write_students(students)
        print("Student added successfully!")

def view_students():
    students = read_students()
    if not students:
        print("No records found.")
        return

    for s in students:
        print(f"ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}")

        
def search_student():
    students = read_students()
    sid = input("Enter ID to search: ")
    for s in students:
        if s["id"] == sid:
            print(f"Found -> ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}")
            return

    print("Student not found.")


def update_student():
    students = read_students()
    sid = input("Enter ID to update: ")
    for s in students:
        if s["id"] == sid:
            s["name"] = input("Enter new name: ")
            s["marks"] = float(input("Enter new marks: "))
            write_students(students)
            print("Student updated successfully!")
            return

    print("Student not found.")


def delete_student():
    students = read_students()
    sid = input("Enter ID to delete: ")
    new_students = [s for s in students if s["id"] != sid]
    if len(new_students) == len(students):
        print("Student not found.")
    else:
        write_students(new_students)
        print("Student deleted successfully!")

        
def show_average():
    students = read_students()
    if not students:
        print("No data available.")
        return
    avg = sum(s["marks"] for s in students) / len(students)
    print(f"Average Marks = {avg:.2f}")


def show_topper():
    students = read_students()
    if not students:
        print("No data available.")
        return
    topper = max(students, key=lambda x: x["marks"])
    print(f"Topper -> ID: {topper['id']}, Name: {topper['name']}, Marks: {topper['marks']}")

def main():
    while True:
        print("\n===== Student Management System =====")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Show Average Marks")
        print("7. Show Topper")
        print("8. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            show_average()
        elif choice == "7":
            show_topper()
        elif choice == "8":
            print("Exiting program...")
            break
        else:
            print("Invalid choice!")
# Call appropriate functions here
main()

```
